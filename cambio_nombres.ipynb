# Cambiar nombre de los archivos del dataset

> Input
*   Directorio que contiene los archivos

> Output
* Carpeta 'dataset_rename' con archivos renombrados


```python
#parametros iniciales necesarios

directorio = '/content/drive/MyDrive/PT/Mask_R-CNN/Mask_RCNN-master/datasets/train'    #directorio que contiene las imagenes y el archivo de etiquetado en formato json

#carpeta = 'dataset_rename'    #carpeta donde se almacenara el nuevo etiquetado generado     
```

```python
#importacion de bibliotecas necesarias

import os
import errno
```

```python
#definicion de variables 

directorio = directorio.replace('\\', '/') + "/"    #se modifica el directorio para correcta lectura de python
contenido_directorio = os.listdir(directorio)  
ejpg = '.jpg'
exml = '.xml'
```

```python
#comprobacion de archivos en directorio

try:
    #os.mkdir(directorio + carpeta)    #se crea una carpeta para almacenar todos los archivos de etiquetado creados

    for archivo in contenido_directorio:
        if os.path.isfile(os.path.join(directorio, archivo)):
            if archivo.endswith(ejpg):
                print("Archivo JPG encontrado: ", directorio + archivo)
                rename = directorio + archivo[5:9] + ejpg
                print("Archivo JPG renombrado: ", rename)
                os.rename((directorio + archivo), rename)

            if archivo.endswith(exml):
                print("Archivo XML encontrado: ", directorio + archivo)
                rename = directorio + archivo[5:9] + exml
                print("Archivo XML renombrado: ", rename)
                os.rename((directorio + archivo), rename)

except OSError as e:
    if e.errno != errno.EEXIST:
        raise
```
